```python
import sys
import fileinput
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np

data = []

for line in fileinput.input('/home/jayant/vision/PETA/TownCentre/archive/Label.txt'):
    data.append((int(line[:-2].split(' ')[0]),line[:-2].split(' ')))
```

```python
atr = []

for line in fileinput.input('attributes.txt'):
    atr.append(line[:-1])
```

```python
data.sort()
```

```python
file2 = open('/home/jayant/vision/PETA/TownCentre_atr.txt','w')
```

```python
for item in data:
    file2.write(item[1][0])
    file2.write(' ')
    for j in atr:
        if j in item[1]:
            file2.write('1')
            file2.write(' ')
        else:
            file2.write('0')
            file2.write(' ')
    file2.write('\n')
    
file2.close()
```
